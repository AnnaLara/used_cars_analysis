```python
listings_info = []

for listing in listings:
        headline = listing.find_element_by_class_name('mt-CardAd-titleHiglight').text
        listing_id = listing.get_attribute("id") 
        
        #unpack attributes
        attributes = listing.find_elements_by_class_name('mt-CardAd-attribute')
        
        attr_list = []
        for attribute in attributes:
            attr_list.append(attribute.text)
            
        try:
            year = int(attr_list[2])
        except:
            pass
        
        try:
            location = attr_list[0]
        except:
            pass
        
        try:
            fuel_type = attr_list[1]
        except:
            pass
        
        #extract km integer from string
        try:
            km = attr_list[3]
            km = km.replace('.', '')
            km = int(km[:-3])
        except:
            pass
        
        price = []
        price_element = listing.find_elements_by_class_name('mt-AdPrice-amount')
        for element in price_element:
            price.append(element.text) 
        
        print(price[0])
        
        #for char in price:
            #if int(char):
                  #price_int.append(char) 
        
        
        #listing_url = 
        #km = 

        listings_info.append({
            'id': listing_id,
            'headline': headline,
            'attributes': attr_list,
            'price': price,
            'year': year,
            'location': location,
            'fuel_type': fuel_type,
            'km': km
        })
```
